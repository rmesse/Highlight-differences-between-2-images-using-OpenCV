### Images rendered using OpenCV

````python
img1 = cv2.imread('images/img1.png')
img2 = cv2.imread('images/img2.png')
````

I convert the images into grayscale (only 1 channel), which is easier to process compared with coloured images (3 channels). 

````python
# Convert the images to grayscale
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
````

The **structural similarity index measure (SSIM)** is a method for predicting the perceived quality of pictures. It is also frequently used for measuring the similarity between two images, which is our intent. The lower the SSIM score, the lower is the similarity.

````python
# Compute the Structural Similarity Index (SSIM) between the two images, ensuring that the difference image is returned
(score, diff) = structural_similarity(img1_gray, img2_gray, full=True)
diff = (diff * 255).astype("uint8")
print("SSIM: {}".format(score))
````

I now use OpenCV's ```threshold``` function to convert an image into a binary format. This function analyzes a grayscale image (```diff```) and applies an adaptive thresholding technique, which dynamically determines the threshold value for each region of the image based on its pixel intensity. This results in a binary image, where pixels are either black (0) or white (255), depending on whether they fall below or above the calculated threshold.

After thresholding, I use the ```findContours``` function to extract the contours from the binary image. Contours are simply curves or boundaries that connect continuous points with the same intensity. They are useful for detecting and analyzing shapes within the image, which can help with tasks like object recognition or tracking.

Finally, the ```grabContours``` function is used to select the appropriate contours from the output of ```findContours```, simplifying the result based on the retrieval mode used. This step ensures that we only keep the necessary contour data for further analysis.

````python
# Calculating threshold and contours 
thresh = cv2.threshold(diff, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
````

The code below loops over the detected contours and draws bounding rectangles around each contour in both images.

````python
# loop over the contours
for c in cnts:
    # applying contours on image
    (x, y, w, h) = cv2.boundingRect(c)
    cv2.rectangle(img1, (x, y), (x + w, y + h), (255, 0, 0), 1)
    cv2.rectangle(img2, (x, y), (x + w, y + h), (255, 0, 0), 1)
````

### Summary 

This approach, combining SSIM, thresholding, and contour detection, provided a comprehensive method for detecting differences between two images by focusing on both global structural similarity and localized shape differences.
